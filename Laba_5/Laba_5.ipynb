```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Загрузка данных
passengers = pd.read_csv('/content/drive/MyDrive/passengers.csv', index_col='Month', parse_dates=True)
def calculate_acf(series, lags):
    acf_values = []
    for lag in range(1, lags + 1):
        shifted = series.shift(lag)
        corr = series.corr(shifted)
        acf_values.append(corr)
    return acf_values

lags = 50
acf_values = calculate_acf(passengers['#Passengers'], lags)
plt.figure(figsize=(10, 6))
plt.plot(range(1, lags + 1), acf_values)
plt.xlabel('Lag')
plt.ylabel('ACF')
plt.show()
```

```python
# birds =/content/drive/MyDrive/passengers.csv
from statsmodels.graphics.tsaplots import plot_acf
plot_acf(passengers['#Passengers'], lags=50)
plt.show()
```

```python
births = pd.read_csv('/content/drive/MyDrive/births.csv', parse_dates=True)
```

```python
acf_values_births = calculate_acf(births['Births'], lags)
plt.figure(figsize=(10, 6))
plt.plot(range(1, lags + 1), acf_values_births)
plt.xlabel('Lag')
plt.ylabel('ACF')
plt.title('ACF for Births Dataset')
plt.show()
```

```python
plot_acf(births['Births'], lags=50)
plt.show()
```

```python
x = np.arange(0, 50, 0.1)
y = np.sin(x)
acf_values_sin = calculate_acf(pd.Series(y), 100)
plt.figure(figsize=(10, 6))
plt.plot(range(1, 101), acf_values_sin)
plt.xlabel('Lag')
plt.ylabel('ACF')
plt.show()
```

```python
plot_acf(pd.Series(y), lags=100)
plt.show()
```

```python
from statsmodels.tsa.stattools import adfuller

result = adfuller(births['Births'])
print('ADF:', result[0])
print('p-value:', result[1])
```
